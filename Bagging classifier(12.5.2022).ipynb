```python
import pandas as pd
file=pd.read_csv("C:/Users/ASUS/Downloads/Social_Network_Ads.csv")
df=pd.DataFrame(file)
df
```

```python
df=pd.get_dummies(df)
df
```

```python
x=df.drop(['Purchased'],axis=1)
x
```

```python
y=df["Purchased"].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)

x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_classifier = BaggingClassifier(
      DecisionTreeClassifier(class_weight='balanced'),
    max_samples=0.5, max_features=0.5, bootstrap=False
)

bag_classifier.fit(x_train, y_train)       #fit code is used here
```

```python
y_pred=bag_classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
classification_report(y_test,y_pred)
```
