```python

Q1 = df.quantile(0.25)


Q3 = df.quantile(0.75)

# Interquartile Range (IQR) is defined as the difference between the third and first quartile
# calculate IQR
IQR = Q3 - Q1

# print the IQR
print(IQR)
```

```python
df = df[~((df< (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

df = df.reset_index(drop = True)
```

```python
# plot multiple boxplots
df.boxplot()

# set plot label
# set text size using 'fontsize'
plt.title('Distribution of Independent Variables', fontsize = 15)

# xticks() returns the x-axis ticks
# 'rotation = vertical' rotates the x-axis labels vertically
plt.xticks(rotation = 'vertical', fontsize = 15)

# display the plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print("x_train",x_train.shape)
print("y_train",y_train.shape)

# print dimension of test set
print("x_test",x_test.shape)
print("y_test",y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classification = KNeighborsClassifier(n_neighbors = 3)
knn_model = knn_classification.fit(x_train, y_train)
```

```python
y_predict=knn_model.predict(x_test)
knn_model.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

```python

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report
z=classification_report(y_test,y_predict)
print(z)
```
